## Project Proposal

### Problem Statement

How can we use collected medical data on children with identified disorders to predict the conditions of children without identified disorders? What are the key features that will have a significant impact in predicting a given disorder?

### Context

With population growing at an exponential rate, there is a need to be able to identify genetic disorders though clinical data and treat it early in children before it develops into something more detrimental. A government agency has provided medical information on childiren with genetic disorders and childrens without identified genetic disorders. Using machine learning, the government agency hopes to provide hosptials and medical doctors the ability to potentially identify genetic disorders to either provide the proper medicare or narrow clincial diagnotics.

### Criteria for Success

Identify the key features in predicting a genetic disorder.
Identify the genetic disorder and disorder subclass in the test data set for each of the patient.

### Scope of Solution Space

Scope is to be only around predicting genetic disorder and disorder subclass.

### Constraints

There are a large variety of genetic disorders and disorder subclass that will not be covered within the data set.
There is a potential for misdiagnosis given the disorder is outside the model's prediction.
Given a large data set that continues to grow, speed of the model and handling of the data will become increasingly costly and slower.

### Stakeholders

Government agency
Hosptial directors
Medial professionals

### Data Sources

Test and Train data sets provided by the government agency as two csv files.
